## 1. Ausgangssituation

Forex-Märkte sind sehr dynamisch und werden von vielen Faktoren beeinflusst: technischen Marktbewegungen, Volatilität, Zinsdifferenzen, makroökonomischen Nachrichten und Marktpsychologie.

Viele Trading-Tools zeigen Charts und Indikatoren, erklären aber nicht immer verständlich, wie Risiko, Positionsgröße, Chance/Risiko-Verhältnis und Szenarioeinschätzung zusammenhängen.

Daraus entsteht die Idee, eine Streamlit-App zu entwickeln, die Forex-Marktdaten analysiert und die Ergebnisse verständlich sowie risikoorientiert aufbereitet.

## 2. Problemstellung

Private Trader und Lernende im Bereich Finanzmärkte betrachten häufig einzelne Charts oder Indikatoren isoliert. Dadurch besteht die Gefahr, dass Handelsentscheidungen nicht ausreichend datenbasiert und risikoorientiert getroffen werden.

Ein zentrales Problem ist, dass technische Analyse, maschinelles Lernen und Risikomanagement oft getrennt voneinander betrachtet werden. ForexScope AI soll diese Perspektiven in einer Anwendung zusammenführen.

## 3. Zielgruppe

Die App richtet sich an:

- Studierende im Bereich Data Analytics und Big Data
- Personen, die Forex-Marktdaten explorativ analysieren möchten
- Lernende, die technische Indikatoren und Risikomanagement verstehen wollen
- Nutzer, die eine transparente Szenarioeinschätzung statt einer einfachen Buy/Sell-Empfehlung benötigen

Wichtig ist: Die App ist keine Finanzberatung und keine Handelsaufforderung.

## 4. Projektidee

**ForexScope AI** ist eine ML-gestützte Streamlit-App zur Analyse von Forex-Marktdaten.

Die App soll:

- Forex-Paare wie EUR/USD, GBP/USD oder USD/JPY analysieren
- verschiedene Timeframes bis hin zu M1 berücksichtigen
- technische Indikatoren berechnen
- Support- und Resistance-Zonen darstellen
- eine ML-basierte Szenarioeinschätzung liefern
- Risiko, Positionsgröße und Break-even-Trefferquote berechnen
- eine spätere Broker-Automation methodisch vorbereiten, aber noch nicht live ausführen

## 11. Grenzen und Risiken

Die App hat bewusst klare Grenzen:

- Sie ist keine Finanzberatung.
- Sie gibt keine sicheren Gewinnversprechen.
- Historische Daten garantieren keine zukünftigen Bewegungen.
- yfinance-Daten können verzögert oder unvollständig sein.
- ML-Prognosen sind nur Szenarioeinschätzungen.
- Eine echte Broker-Automation wäre erst nach Paper-Trading, Logging, Risk Engine und Kill Switch vertretbar.

Diese Grenzen sind wichtig, damit das Projekt seriös bleibt.
