## 1. Q-Phase: Ausgangsfrage und Problemrahmen

Forex-Märkte sind dynamisch, volatil und stark von technischen, makroökonomischen und psychologischen Faktoren beeinflusst.

Viele Tools zeigen zwar Charts und Indikatoren, verbinden diese aber nicht ausreichend mit Risikomanagement, Positionsgröße und Modellunsicherheit.

Daraus ergibt sich die Projektidee: Eine Streamlit-App, die technische Analyse, ML-Szenario und Capital/Risk Planning zusammenführt.

## 2. Zentrale Fragestellung

**Wie können historische Forex-Marktdaten genutzt werden, um eine ML-gestützte Streamlit-App zu entwickeln, die technische Marktanalyse, Szenarioeinschätzung und risikobasierte Positionsplanung verständlich kombiniert?**

Diese Fragestellung verbindet:

1. Datenanalyse historischer Forex-Marktdaten  
2. technische Indikatoren und Multi-Timeframe-Analyse  
3. ML-basierte Szenarioeinschätzung  
4. risikobasierte Positionsplanung  
5. transparente Kommunikation von Grenzen und Unsicherheit

## 3. Zielgruppe

Die App richtet sich an:

- Studierende im Bereich Data Analytics und Big Data
- Personen, die Forex-Marktdaten explorativ analysieren möchten
- Lernende, die technische Analyse und Risikomanagement verstehen wollen
- Nutzer, die keine einfache Buy/Sell-Empfehlung, sondern eine transparente Szenarioeinschätzung benötigen

Die App ist keine Finanzberatung und keine Handelsaufforderung.

## 5. A-Phase: Analyse- und Modellierungslogik

ForexScope AI berechnet aus den Marktdaten mehrere Analysebausteine:

### Technische Analyse

- Moving Averages
- RSI
- MACD
- ATR
- Support und Resistance
- Multi-Timeframe-Score

### Machine Learning

Ein Random-Forest-Modell wird genutzt, um eine einfache bullish/bearish-Szenarioeinschätzung für die nächsten Kerzen zu erstellen, sofern genügend Daten vorhanden sind.

### Capital & Risk Planning

Die App berechnet:

- Risiko pro Trade
- Stop-Loss-Abstand in Pips
- Take-Profit-Abstand in Pips
- Chance/Risiko-Verhältnis
- Break-even-Trefferquote
- Positionsgröße in Units/Lots

## 6. KPI Framework

Die App wird anhand mehrerer KPI-Gruppen bewertet.

### Daten-KPIs

| KPI | Bedeutung |
|---|---|
| Datenpunkte | Anzahl verfügbarer OHLC-Kerzen |
| Timeframe-Abdeckung | unterstützte Timeframes von M1 bis W1 |
| Datenaktualität | Aktualität der geladenen Marktdaten |
| Datenverfügbarkeit | ob Daten für ein Forex-Paar geladen werden können |

### Analyse-KPIs

| KPI | Bedeutung |
|---|---|
| Trend-Score | zusammengefasste technische Markteinschätzung |
| ATR in Pips | Volatilitätsmaß |
| Support-Abstand | Abstand zum nächsten Support |
| Resistance-Abstand | Abstand zum nächsten Widerstand |
| Multi-Timeframe-Score | Gesamtbild über mehrere Timeframes |

### ML-KPIs

| KPI | Bedeutung |
|---|---|
| Accuracy | Modellgüte auf Testdaten |
| Trainingsdatenpunkte | verfügbare Datenmenge für ML |
| Szenario-Wahrscheinlichkeit | Wahrscheinlichkeit für bullish/bearish |
| Modellverfügbarkeit | ob genug Daten für ML vorhanden sind |

### Risiko-KPIs

| CRV | Chance/Risiko-Verhältnis |

## 7. 3-Phase: Drei zentrale Projektergebnisse

### Ergebnis 1: Live Forex Analyse

Die App zeigt aktuelle Forex-Daten, Candlestick-Charts, technische Indikatoren, Support/Resistance und Trend-Score.

### Ergebnis 2: Multi-Timeframe Dashboard

Die App analysiert mehrere Timeframes und verdichtet diese zu einer übergeordneten Einschätzung.

Ziel ist nicht, Nutzer zu einem Trade zu bewegen, sondern Daten, Risiko und Unsicherheit verständlich zu machen.

## 9. Grenzen und Risiken

Die App hat bewusst klare Grenzen:

- keine Finanzberatung
- keine sicheren Gewinnversprechen
- keine Live-Trading-Automation
- yfinance-Daten können verzögert oder unvollständig sein
- Forex-Volumen ist über Yahoo Finance nur eingeschränkt nutzbar
- ML-Prognosen sind Szenarioeinschätzungen, keine sicheren Vorhersagen
- Broker-Automation wäre erst nach Paper-Trading, Logging, Risk Engine und Kill Switch vertretbar
