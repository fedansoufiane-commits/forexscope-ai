# 04 – C: Conclude and Compare

## Ergebnisse bewerten und vergleichen

**QUA3CK-Phase:** C – Conclude and compare  

In dieser Phase werden die Ergebnisse der Modellierung und App-Analyse bewertet. Es geht darum, Schlussfolgerungen zu ziehen, Modelle und Kennzahlen zu vergleichen und zu prüfen, ob die Ergebnisse zur ursprünglichen Fragestellung passen.

Für ForexScope AI bedeutet das:

- technische Analyse bewerten,
- ML-Ergebnisse einordnen,
- Risiko-KPIs interpretieren,
- Timeframes vergleichen,
- Grenzen der Ergebnisse transparent machen.

## 1. Bewertungslogik

ForexScope AI bewertet nicht nur ein ML-Modell, sondern ein gesamtes Analyseprodukt.

Bewertet werden:

1. Datenqualität,
2. technische Analyse,
3. ML-Szenario,
4. Risiko-Kennzahlen,
5. Nutzbarkeit der App,
6. Transparenz der Grenzen.

Dadurch wird die App nicht auf Accuracy reduziert.

## 2. Vergleichsebenen

Die App vergleicht Ergebnisse auf mehreren Ebenen.

| Vergleich | Frage |
|---|---|
| Timeframe-Vergleich | Stimmen kurzfristige und langfristige Signale überein? |
| Indikator-Vergleich | Bestätigen RSI, MACD und Moving Averages einander? |
| Modell-Vergleich | Wie gut ist das ML-Szenario im Test? |
| Risiko-Vergleich | Ist das CRV sinnvoll? |
| Datenvergleich | Sind genug Daten vorhanden? |

Diese Vergleichslogik macht die Schlussfolgerung robuster.

## 3. KPIs für Conclude and Compare

| KPI-Gruppe | Beispiele |
|---|---|
| Daten-KPIs | Datenpunkte, Missing Values, Aktualität |
| Analyse-KPIs | Trend-Score, ATR, Support/Resistance |
| ML-KPIs | Accuracy, Trainingsdaten, Szenario-Wahrscheinlichkeit |
| Risiko-KPIs | CRV, Risiko pro Trade, Break-even-Trefferquote |
| App-KPIs | Interpretierbarkeit, Präsentationsfähigkeit, Transparenz |

Die zentrale Bewertung lautet nicht: „Ist der Trade sicher?“  
Sondern: „Sind Daten, Szenario und Risiko nachvollziehbar genug dargestellt?“

## 4. Interpretation des ML-Ergebnisses

Das ML-Ergebnis wird bewusst als Szenarioeinschätzung interpretiert.

Ein Ergebnis wie „bullish mit 58 % Wahrscheinlichkeit“ bedeutet nicht, dass der Kurs sicher steigt. Es bedeutet lediglich, dass das Modell auf Basis der historischen Features in dieser Situation eher eine Aufwärtsbewegung klassifiziert.

Daraus folgt:

- Das ML-Ergebnis ist ein Analysehinweis.
- Es ist keine Handelsentscheidung.
- Es muss mit technischer Analyse und Risiko-KPIs kombiniert werden.
- Es muss bei zu wenig Daten unterdrückt oder als nicht verfügbar angezeigt werden.

## 5. Interpretation der Risiko-KPIs

Besonders wichtig ist das Chance/Risiko-Verhältnis.

Beispiel:

| CRV | Break-even-Trefferquote |
|---:|---:|
| 1.0 | 50.0 % |
| 1.5 | 40.0 % |
| 2.0 | 33.3 % |
| 3.0 | 25.0 % |

Diese Werte zeigen: Eine Trefferquote ist nicht bekannt, sondern muss später aus Backtests oder einem Trading-Journal geschätzt werden. Die App fragt deshalb keine Trefferquote als Pflichtwert ab, sondern berechnet zuerst, welche Trefferquote rechnerisch nötig wäre.

## 6. Schlussfolgerung zur App

ForexScope AI ist erfolgreich, wenn:

- Daten geladen und geprüft werden,
- technische Indikatoren nachvollziehbar dargestellt werden,
- ML nur bei ausreichender Datenbasis genutzt wird,
- Risiko-KPIs verständlich berechnet werden,
- Grenzen klar kommuniziert werden,
- Nutzer eine bessere Entscheidungsgrundlage erhalten.

Die App ist nicht erfolgreich, wenn sie scheinbar sichere Signale erzeugt oder Risiken verschleiert.

## 7. Grenzen des Vergleichs

Die C-Phase muss Grenzen offen benennen:

- yfinance ist keine professionelle Brokerdatenquelle,
- historische Daten garantieren keine Zukunft,
- Intraday-Daten sind begrenzt,
- Accuracy allein ist keine ausreichende Erfolgsmetrik,
- technische Indikatoren können Fehlsignale liefern,
- echte Broker-Automation braucht Paper-Trading, Logging und Sicherheitsmechanismen.

Diese Grenzen sind kein Schwachpunkt, sondern Teil einer seriösen Bewertung.

## 8. Ergebnis der C-Phase

Die C-Phase zeigt:

- ForexScope AI kann technische Analyse, ML und Risiko-KPIs sinnvoll verbinden.
- Die Ergebnisse müssen vergleichend und vorsichtig interpretiert werden.
- KPIs schaffen Transparenz.
- Die App ist als Analyse- und Bildungswerkzeug geeignet.
- Für produktive Automation sind weitere Schritte notwendig.

Damit wird die Grundlage für die K-Phase geschaffen: den Wissenstransfer in eine nutzbare Lösung.
