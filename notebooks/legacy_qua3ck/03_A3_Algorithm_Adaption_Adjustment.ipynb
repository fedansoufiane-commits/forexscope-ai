# 03 – A³: Algorithm Selection, Adaption, Adjustment

## Der iterative Modellierungsprozess von ForexScope AI

**QUA3CK-Phase:** A³  
**Bestandteile:** Algorithm selection, Adaption, Adjustment  

Die A³-Phase ist der iterative Kern des QUA3CK-Prozesses. Hier werden Algorithmen ausgewählt, Daten angepasst und Hyperparameter optimiert.

Für ForexScope AI bedeutet das:

1. geeignete Algorithmen für Forex-Zeitreihendaten auswählen,
2. Rohdaten in ML-fähige Features umwandeln,
3. Modellparameter und Timeframe-Logik anpassen,
4. Ergebnisse wiederholt prüfen und verbessern.

## 1. A³ als iterative Schleife

Der A³-Prozess ist keine lineare Einmalentscheidung. Er wird wiederholt, bis ein fachlich und technisch sinnvoller Modellstand erreicht ist.

Die Schleife lautet:

1. **Algorithm Selection:** Welcher Algorithmus eignet sich?
2. **Adaption:** Wie müssen Daten und Features angepasst werden?
3. **Adjustment:** Welche Parameter müssen justiert werden?

Danach wird geprüft, ob die Ergebnisse plausibel sind. Falls nicht, beginnt die Schleife erneut.

## 2. Algorithm Selection

Für ForexScope AI kommen verschiedene Modellansätze in Frage.

| Algorithmus | Vorteil | Grenze |
|---|---|---|
| Logistic Regression | einfach interpretierbar | begrenzte Nichtlinearität |
| Decision Tree | gut erklärbar | Overfitting-Risiko |
| Random Forest | robust, nichtlinear, für Prototyp geeignet | weniger transparent als einfache Modelle |
| Gradient Boosting | leistungsfähig | komplexer |
| LSTM/Neuronale Netze | geeignet für Sequenzen | hoher Daten- und Komplexitätsbedarf |

Für den Prototyp wird ein **Random Forest Classifier** gewählt.

Begründung:

- geeignet für tabellarische Features,
- robust gegenüber nichtlinearen Beziehungen,
- relativ schnell trainierbar,
- verständlich genug für ein Studienprojekt,
- besser geeignet als ein zu komplexes neuronales Netz im frühen Prototyp.

## 3. Zielvariable

Das Modell formuliert ein binäres Klassifikationsproblem.

Die Zielvariable lautet:

- **1 = bullish / Kurs liegt nach definiertem Horizont höher**
- **0 = bearish oder neutral / Kurs liegt nach definiertem Horizont nicht höher**

Die App modelliert damit keine exakte Preisprognose, sondern eine Szenarioeinschätzung.

## 4. Adaption: Datenanpassung und Feature Engineering

Die Rohdaten müssen in ML-fähige Features umgewandelt werden.

Verwendete Feature-Gruppen:

| Feature-Gruppe | Beispiele |
|---|---|
| Returns | return_1, return_3, return_5, return_10 |
| Trend | Fast MA, Slow MA, MA-Abstände |
| Momentum | MACD, MACD Signal, MACD Histogram |
| Überkauft/Überverkauft | RSI |
| Volatilität | ATR, Volatility 10, Volatility 20 |
| Zielvariable | Future Return, Target |

Die Adaption ist entscheidend, weil der Algorithmus nicht direkt aus rohen Kursen lernen soll, sondern aus strukturierten Merkmalen.

## 5. Timeframe-spezifische Adaption

Ein zentrales Designprinzip ist die Anpassung an unterschiedliche Timeframes.

Beispiel:

| Timeframe | Fast MA | Slow MA | RSI | ATR | Grund |
|---|---:|---:|---:|---:|---|
| M1 | 9 | 21 | 7 | 7 | schnelle Intraday-Reaktion |
| M5 | 9 | 21 | 14 | 14 | kurzfristige Stabilität |
| M15 | 20 | 50 | 14 | 14 | stabilere Struktur |
| H1 | 20 | 50 | 14 | 14 | intraday/mittelfristig |
| D1 | 20 | 50 | 14 | 14 | Tagesanalyse |
| W1 | 20 | 50 | 14 | 14 | langfristige Struktur |

Dadurch wird verhindert, dass ein kurzer Timeframe unpassend mit langfristigen Indikatoren wie SMA 200 überladen wird.

## 6. Adjustment: Hyperparameter-Anpassung

Im Prototyp werden für den Random Forest folgende Parameter genutzt:

| Parameter | Bedeutung |
|---|---|
| n_estimators | Anzahl der Entscheidungsbäume |
| max_depth | maximale Tiefe der Bäume |
| min_samples_leaf | Mindestanzahl Samples pro Blatt |
| class_weight | Ausgleich möglicher Klassenungleichgewichte |
| random_state | Reproduzierbarkeit |

Diese Parameter werden so gewählt, dass das Modell stabil bleibt und nicht zu stark überfitten soll.

## 7. Modell-KPIs

Für die Bewertung innerhalb der A³-Phase werden Modell-KPIs genutzt.

| KPI | Bedeutung |
|---|---|
| Accuracy | Anteil korrekt klassifizierter Testfälle |
| Trainingsdatenpunkte | Datenmenge für Training |
| Testdatenpunkte | Datenmenge für Validierung |
| Szenario-Wahrscheinlichkeit | Wahrscheinlichkeit für bullish/bearish |
| Modellverfügbarkeit | ob genug Daten für Training vorhanden sind |

Wichtig ist: Wenn zu wenig Daten vorhanden sind, gibt die App kein künstlich präzises ML-Ergebnis aus.

## 8. Iterative Verbesserung

Mögliche Iterationen im A³-Prozess:

1. weitere Features hinzufügen,
2. andere Algorithmen testen,
3. Hyperparameter systematisch optimieren,
4. Timeframe-Konfigurationen anpassen,
5. externe Kontextdaten integrieren,
6. Backtesting ergänzen,
7. Paper-Trading-Daten als Feedback nutzen.

Damit bleibt der A³-Prozess offen für Weiterentwicklung.

## 9. Ergebnis der A³-Phase

Die A³-Phase liefert:

- einen begründeten Startalgorithmus,
- ein Feature-Engineering-Konzept,
- eine timeframe-spezifische Anpassungslogik,
- erste Hyperparameter,
- KPIs zur Modellbewertung,
- eine klare Grundlage für Vergleich und Schlussfolgerung in der C-Phase.

ForexScope AI erfüllt damit den iterativen Kern des QUA3CK-Modells.
