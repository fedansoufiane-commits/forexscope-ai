## ForexScope AI – Data-Driven Forex Insights

**QUA3CK-Phase:** Q – Question  
**Projekt:** ForexScope AI  
**App-Typ:** ML-gestützte Streamlit-App  
**Domäne:** Forex-Marktdaten, technische Analyse, Machine Learning, Risikomanagement  

Die Q-Phase definiert das Machine-Learning-Problem. Sie beantwortet nicht direkt, welcher Algorithmus verwendet wird, sondern zuerst, welches fachliche Problem durch eine ML-Anwendung untersucht oder gelöst werden soll.

Ziel dieser Phase ist es, aus einer allgemeinen Idee eine präzise, überprüfbare und für eine App geeignete Fragestellung abzuleiten.

## 1. Ausgangssituation

Forex-Märkte erzeugen fortlaufend große Mengen an Zeitreihendaten. Wechselkurse wie EUR/USD, GBP/USD oder USD/JPY verändern sich laufend und können auf unterschiedlichen Zeitebenen analysiert werden, zum Beispiel M1, M5, H1, D1 oder W1.

Viele bestehende Trading-Tools zeigen Charts und technische Indikatoren. Häufig bleibt aber unklar, wie diese Informationen strukturiert interpretiert werden sollen. Besonders problematisch ist, dass technische Analyse, Machine Learning und Risikomanagement oft getrennt betrachtet werden.

Dadurch entsteht ein fachliches Problem: Nutzer sehen Kursdaten und Indikatoren, erhalten aber keine transparente Verbindung zwischen Marktlage, Modellunsicherheit, Risiko und Positionsgröße.

## 2. Problemdefinition

Das Machine-Learning-Problem von ForexScope AI besteht nicht darin, sichere Gewinne vorherzusagen. Das wäre fachlich unseriös.

Das eigentliche Problem lautet:

> Wie kann aus historischen Forex-Marktdaten eine nachvollziehbare Szenarioeinschätzung abgeleitet werden, die technische Analyse und Risikomanagement verständlich kombiniert?

Das ML-Modell soll daher nicht als automatische Handelsentscheidung verstanden werden, sondern als ein Baustein einer datengetriebenen Entscheidungsunterstützung.

## 3. Projektidee

**ForexScope AI** ist eine Streamlit-App, die Forex-Marktdaten lädt, technische Indikatoren berechnet, ein ML-basiertes Szenario ableitet und Risiko-Kennzahlen verständlich darstellt.

Die App soll:

- Forex-Paare analysieren,
- unterschiedliche Timeframes unterstützen,
- technische Indikatoren wie RSI, MACD, ATR und Moving Averages berechnen,
- Support und Resistance ableiten,
- ein bullish/bearish-Szenario modellieren,
- Risiko, Positionsgröße und Break-even-Trefferquote berechnen,
- Ergebnisse transparent und lernorientiert kommunizieren.

## 4. Zielgruppe

Die Zielgruppe sind nicht professionelle Trader mit Live-Broker-Infrastruktur, sondern Lernende und dateninteressierte Nutzer.

| Zielgruppe | Nutzen der App |
|---|---|
| Studierende | Anwendung eines ML-Prozessmodells an einem realitätsnahen Projekt |
| Data-Analytics-Lernende | Verständnis von Zeitreihendaten und Feature Engineering |
| Forex-Interessierte | strukturierte Analyse von Forex-Paaren |
| Nutzer technischer Analyse | Verbindung von Indikatoren, Szenario und Risiko |
| App-Entwickler | Beispiel für eine Streamlit-basierte ML-Anwendung |

Die App ist ausdrücklich keine Finanzberatung und keine Handelsaufforderung.

## 5. Zentrale Forschungs- und Projektfrage

Die zentrale Frage der Q-Phase lautet:

> **Wie können historische Forex-Marktdaten genutzt werden, um eine ML-gestützte Streamlit-App zu entwickeln, die technische Marktanalyse, Szenarioeinschätzung und risikobasierte Positionsplanung verständlich kombiniert?**

Diese Frage ist für das Projekt geeignet, weil sie:

- eine klare Datenbasis nennt,
- eine ML-Komponente enthält,
- einen praktischen App-Bezug hat,
- Risikomanagement integriert,
- keine unrealistische sichere Kursprognose verspricht.
