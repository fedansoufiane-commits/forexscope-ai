# 05 – K: Knowledge Transfer

## Wissenstransfer, Produktisierung und Reflexion

**QUA3CK-Phase:** K – Knowledge transfer  

In der K-Phase wird aus Analyse und Modellierung eine nutzbare Lösung, ein Produkt, ein Prototyp oder eine dokumentierte Erkenntnis.

Für ForexScope AI bedeutet das:

- Umsetzung als Streamlit-App,
- Dokumentation im GitHub-Repository,
- Präsentation entlang der QUA3CK-Phasen,
- Vorbereitung eines Handouts,
- Reflexion von Grenzen und Ausblick.

## 1. Wissenstransfer in eine App

ForexScope AI überträgt die Erkenntnisse aus den vorherigen Phasen in eine interaktive Streamlit-App.

Die App macht die Analyse praktisch nutzbar:

- Forex-Paar auswählen,
- Timeframe wählen,
- Marktdaten laden,
- technische Analyse anzeigen,
- ML-Szenario prüfen,
- Risiko und Positionsgröße berechnen,
- Automationsfähigkeit reflektieren.

Damit wird aus einem ML-Konzept ein bedienbarer Prototyp.

## 2. Repository als Wissensträger

Das GitHub-Repository dient als strukturierte Projektdokumentation.

Es enthält:

| Bestandteil | Zweck |
|---|---|
| app.py | lauffähige Streamlit-App |
| requirements.txt | technische Reproduzierbarkeit |
| README.md | Projektübersicht |
| docs/data_sources.md | Datenquellen-Dokumentation |
| docs/kpi_framework.md | KPI-System |
| docs/qua3ck_process.md | Prozessdokumentation |
| notebooks/ | QUA3CK-Phasen als Jupyter Notebooks |

Dadurch ist das Projekt nachvollziehbar, reproduzierbar und präsentierbar.

## 3. Kommunikation in der Präsentation

Die Präsentation sollte entlang der QUA3CK-Logik aufgebaut werden:

1. **Q:** Welche Frage beantwortet die App?
2. **U:** Welche Daten werden genutzt und verstanden?
3. **A³:** Welche Algorithmen, Features und Parameter werden verwendet?
4. **C:** Wie werden Ergebnisse bewertet und verglichen?
5. **K:** Wie wird das Wissen in eine App und Dokumentation übertragen?

Damit entsteht ein roter Faden vom Problem bis zum Prototyp.

## 4. Wissenstransfer für Nutzer

Die App vermittelt Nutzern Wissen über:

- Forex-Zeitreihendaten,
- technische Indikatoren,
- ML-Szenarien,
- Risiko pro Trade,
- Positionsgrößen,
- Break-even-Trefferquote,
- Grenzen datengetriebener Prognosen.

Der wichtigste Wissenstransfer besteht darin, dass Nutzer nicht nur ein Signal sehen, sondern verstehen, wie Daten, Modell und Risiko zusammenhängen.

## 5. Grenzen und Verantwortung

Da Forex ein risikobehaftetes Thema ist, muss der Wissenstransfer verantwortungsvoll erfolgen.

Die App vermeidet:

- Gewinnversprechen,
- aggressive Trading-Sprache,
- automatische Kauf-/Verkaufsempfehlungen,
- Verschleierung von Modellunsicherheit,
- Live-Trading ohne Sicherheitsarchitektur.

Stattdessen nutzt sie:

- Warnhinweise,
- Methodik-Erklärungen,
- Risiko-KPIs,
- klare Abgrenzung von Finanzberatung,
- Automation Readiness nur als Ausblick.
