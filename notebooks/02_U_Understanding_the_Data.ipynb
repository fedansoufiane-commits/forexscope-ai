## 1. Fachliches Datenverständnis

Forex-Daten beschreiben Wechselkurse zwischen zwei Währungen. Ein Währungspaar besteht aus:

- Basiswährung,
- Kurswährung.

Beispiel:

- EUR/USD = Anzahl US-Dollar für 1 Euro.
- GBP/USD = Anzahl US-Dollar für 1 britisches Pfund.
- USD/JPY = Anzahl japanischer Yen für 1 US-Dollar.

Die App nutzt diese Kurse als Zeitreihe. Jede Kerze repräsentiert einen Zeitraum, beispielsweise eine Minute, fünf Minuten, eine Stunde oder einen Tag.

## 2. Datenstruktur: OHLC-Daten

Die wichtigste Datenstruktur sind OHLC-Daten.

| Spalte | Bedeutung |
|---|---|
| Open | Eröffnungskurs der Kerze |
| High | höchster Kurs im Zeitraum |
| Low | niedrigster Kurs im Zeitraum |
| Close | Schlusskurs der Kerze |
| Date/Datetime | Zeitstempel |
| Volume | bei Forex über Yahoo meist eingeschränkt aussagekräftig |

Für ForexScope AI sind vor allem Open, High, Low und Close relevant. Aus ihnen werden technische Indikatoren und ML-Features berechnet.

## 3. Unterstützte Timeframes

Die App soll mehrere Zeitebenen betrachten.

| Timeframe | Bedeutung | Herausforderung |
|---|---|---|
| M1 | 1-Minuten-Kerzen | sehr viel Rauschen, begrenzte Historie |
| M5 | 5-Minuten-Kerzen | kurzfristige Intraday-Analyse |
| M15 | 15-Minuten-Kerzen | stabilere Intraday-Struktur |
| M30 | 30-Minuten-Kerzen | Übergang zu Stundenanalyse |
| H1 | Stundenkerzen | gutes Intraday-Gleichgewicht |
| H4 | mehrstündige Struktur | mittelfristige Perspektive |
| D1 | Tageskerzen | übergeordneter Trend |
| W1 | Wochenkerzen | langfristige Perspektive |

Kurze Timeframes benötigen andere Indikatorlängen als lange Timeframes. Daher arbeitet ForexScope AI mit dynamischen Timeframe-Konfigurationen.

## 5. Datenqualitätsprobleme

Forex-Daten können verschiedene Qualitätsprobleme aufweisen:

- fehlende Werte,
- begrenzte historische Intraday-Daten,
- abweichende Daten je Anbieter,
- verzögerte Kurse,
- unklare Zeitzonen,
- Wochenend- und Feiertagslücken,
- kein verlässliches zentrales Forex-Volumen,
- Differenz zwischen Yahoo-Daten und echten Broker-Bid/Ask-Daten.

Die App muss daher transparent anzeigen, wenn zu wenig Daten für bestimmte Analysen oder ML-Modelle vorhanden sind.
